```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

## Binomialverteilung

```python
N = 677117
x = 347716
p = 0.5
```

```python
# P(X>=347716|p=0.5)
1-stats.binom.cdf(x-1,N,p)
```

```python
x = np.arange(0,N)
```

```python
prob = stats.binom.pmf(x,N,p=0.5)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(
    x,
    prob,
    ms=1,
    label=f"Binom(N={N},p={p})"
    )
ax.set_title('Dichtefunktion der Binomialverteilung')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_xlim([330000,350000])
ax.legend()
plt.show()
```

```python
np.argmax(prob)
```

## Einseitiger Test (größer)

```python
N = 2000  # Gesamtzahl der Geburten
x = 1050  # Anzahl der männlichen Geburten
p = 0.5

# Durchführung des Binomialtests
p_value = stats.binomtest(x, N, p, alternative='greater').pvalue

# Ausgabe des p-Werts
print(f'p-Wert: {p_value}')
```

```python
# P(X>=x|p), d.h. Wahrscheinlichkeit, dass die Anzahl der männlichen Geburten größer gleich x ist bei p=0,5,
# d.h. Jungen und Mädchen mit gleicher Wahrscheinlichkeit lebendgeboren
1 - stats.binom.cdf(x-1,N,p)
```

## Einseitiger Test (kleiner)

```python
print(f"p-Wert: {stats.binomtest(x, N, p, alternative='less').pvalue}")
```

```python
# P(X<=x|p), d.h. Wahrscheinlichkeit, dass die Anzahl der männlichen Geburten kleiner gleich x ist bei p=0,5,
# d.h. Jungen und Mädchen mit gleicher Wahrscheinlichkeit lebendgeborenstats.binom.cdf(x,N,p)
```

## Zweiseitger Test (Unterschied)

```python
print(f"p-Wert: {stats.binomtest(x, N, p, alternative='two-sided').pvalue}")
```

```python
# P(X>=x|p), d.h. Wahrscheinlichkeit, dass die Anzahl der männlichen Geburten größer gleich x ist bei p=0,5,
# d.h. Jungen und Mädchen mit gleicher Wahrscheinlichkeit lebendgeboren
upper = (1 - stats.binom.cdf(x-1,N,p))

# P(N-X<=x|p), d.h. Wahrscheinlichkeit, dass die Anzahl der weiblichen Geburten kleiner gleich x ist bei p=0,5,
# d.h. Jungen und Mädchen mit gleicher Wahrscheinlichkeit lebendgeboren
lower = stats.binom.cdf(N-x,N,p)

lower + upper
```
